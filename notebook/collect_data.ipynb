```python
import pandas as pd
import os

csv_list = os.listdir('./data/')
csv_list = ['./data/'+i for i in csv_list if i.endswith('.csv')]

df1 = pd.read_csv(csv_list[0])
df2 = pd.read_csv(csv_list[1])
df3 = pd.read_csv(csv_list[2])
df4 = pd.read_csv(csv_list[3])
```

```python
result = []
```

```python
for index, row in df1.iterrows():
    result.append({
        "emoji": row["Representation"],
        "text_name": row["Name"],
        "unicode": row["CodePoint"],
    })
```

```python
for index, row in df4.iterrows():
    result.append({
        "emoji": row["emoji"],
        "text_name": f"{row['name']}",
        "unicode": row["unicode"],
    })
```

```python
def remove_duplicate_emojis(df):
    seen = {}
    new_rows = []
    
    for _, row in df.iterrows():
        emoji, text_name, unicode_val = row['emoji'], row['text_name'], row['unicode']
        
        if emoji in seen:
            seen[emoji]['text_name']
        else:
            seen[emoji] = {'text_name': text_name, 'unicode': unicode_val}
    
    for emoji, values in seen.items():
        new_rows.append({'emoji': emoji, 'text_name': values['text_name'], 'unicode': values['unicode']})
    
    return pd.DataFrame(new_rows)
```

```python
new_df = remove_duplicate_emojis(df)
```

```python
new_df.to_csv('./data/process_emojis_no_duplicates.csv', index=False)
```
